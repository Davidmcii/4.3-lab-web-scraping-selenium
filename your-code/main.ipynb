#### Download, parse, and print the content from the Trending Developers page from GitHub:

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/developers'
driver.get(url)
````

````python
#your code
nombre=driver.find_element(By.XPATH,'//*[@id="js-pjax-container"]/div[3]').text

nombre
````

````python
caja= driver.find_element(By.XPATH,'//*[@id="js-pjax-container"]/div[3]')

names= [e.text for e in caja.find_elements(By.CLASS_NAME, 'h3.lh-condensed')]

len(names)
````

````python

alias= [e for e in caja.find_elements(By.CLASS_NAME, 'f4.text-normal.mb-1')]


alias[0].find_element(By.TAG_NAME,'a').get_attribute('href')

urls=[]
for el in range(len(alias)):
    url=alias[el].find_element(By.TAG_NAME,'a').get_attribute('href')
    urls.append(url.removeprefix('https://github.com/'))

len(urls)
````

````python
trending=dict(zip(names,urls))

trending
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
#your code


caja=driver.find_elements(By.TAG_NAME,'tbody')

teres=[el for el in caja[4].find_elements(By.TAG_NAME,'tr')]

cada_fila=[]
for i in range(len(teres)):
    temporal=[]
    fila=teres[i].find_elements(By.TAG_NAME,'td')
    for elements in fila: 
        temporal.append(elements.text)
    cada_fila.append(temporal)
    
cada_fila

    

````
